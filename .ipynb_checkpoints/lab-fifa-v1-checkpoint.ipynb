```python
import pandas as pd
import numpy as np
import math

import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline

fifa_df = pd.read_csv('fifa21_train.csv')

fifa_df.shape
```

```python
fifa_df.columns.values
```

```python
fifa_df = fifa_df.drop(['ID', 'Name', 'Club', 'Hits', 'Nationality', 'Position','Team & Contract','Height','Weight', 'foot','Joined','Loan Date End','Wage','Release Clause','Contract'], axis=1)

fifa_df.columns.values
```

```python
# drop duplicates

fifa_df.drop_duplicates()
```

```python
# nan_positions_tuples = list(zip(*np.where(fifa_df.isna())))
# print(nan_positions_tuples)

fifa_df.columns[fifa_df.isna().any()].tolist()
```

```python
# filling NaN values with mean and mode

volleys_mean = fifa_df['Volleys'].mean()
curve_mean = fifa_df['Curve'].mean()
agility_mean = fifa_df['Agility'].mean()
balance_mean = fifa_df['Balance'].mean()
jumping_mean = fifa_df['Jumping'].mean()
interceptions_mean = fifa_df['Interceptions'].mean()
positioning_mean = fifa_df['Positioning'].mean()
visions_mean = fifa_df['Vision'].mean()
composure_mean = fifa_df['Composure'].mean()
sliding_mean = fifa_df['Sliding Tackle'].mean()
aw_mode = fifa_df['A/W'].mode()[0]
dw_mode = fifa_df['D/W'].mode()[0]


fifa_df['Volleys'].fillna(volleys_mean, inplace=True)
fifa_df['Curve'].fillna(curve_mean, inplace=True)
fifa_df['Agility'].fillna(agility_mean, inplace=True)
fifa_df['Balance'].fillna(balance_mean, inplace=True)
fifa_df['Jumping'].fillna(jumping_mean, inplace=True)
fifa_df['Interceptions'].fillna(interceptions_mean, inplace=True)
fifa_df['Positioning'].fillna(positioning_mean, inplace=True)
fifa_df['Vision'].fillna(visions_mean, inplace=True)
fifa_df['Composure'].fillna(composure_mean, inplace=True)
fifa_df['Sliding Tackle'].fillna(sliding_mean, inplace=True)
fifa_df['A/W'].fillna(aw_mode, inplace=True)
fifa_df['D/W'].fillna(dw_mode, inplace=True)

# check again for any NaN values

fifa_df.columns[fifa_df.isna().any()].tolist()
```

```python
# replace ★ values within columns

fifa_df['W/F'].str.replace('★', '')
fifa_df['SM'].str.replace('★', '')
fifa_df['IR'].str.replace('★', '')
```

```python
def sum_values(value_str):
    values = value_str.split('+')
    return sum(map(int, values))

columns_to_change = ['LS', 'ST', 'RS','LW', 'LF', 'CF', 'RF', 'RW', 'LAM','CAM', 'RAM', 'LM', 'LCM','CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM', 'RWB', 'LB', 'LCB','CB', 'RCB', 'RB', 'GK']

fifa_df[columns_to_change] = fifa_df[columns_to_change].applymap(sum_values)
fifa_df[columns_to_change] = fifa_df[columns_to_change].astype(int)
```

```python
def convert_currency (value):
    if 'K' in value:
        return float(value.replace('€', '').replace('K', '')) * 1000
    elif 'M' in value:
        return float(value.replace('€', '').replace('M', '')) * 1000000
    else:
        return float(value.replace('€', ''))
    
fifa_df['Value'] = fifa_df['Value'].apply(convert_currency)

print(fifa_df['Value'])
```

```python
fifa_df.shape
```

```python
fifa_df['BP'] = fifa_df['BP'].replace({'RB':'DEF', 'LB':'DEF', 'CB':'DEF', 'LWB' : 'DEF', 'RWB':'DEF'})
fifa_df['BP'] = fifa_df['BP'].replace({'RW':'MID', 'LW':'MID', 'RM':'MID', 'LM' : 'MID', 'CDM':'MID', 'CAM':'MID','CM':'MID'})
fifa_df['BP'] = fifa_df['BP'].replace({'ST': 'FWD', 'CF': 'FWD'})
```

```python
for column_name, dtype in fifa_df.dtypes.items():
    print("Column:", {column_name}, "Data Type:", {dtype})
```

```python
fifa_df.describe()
```

```python
y = fifa_df['OVA']
X = fifa_df.drop(['OVA'], axis=1)     

X_num = X.select_dtypes(include=np.number)
X_cat = X.select_dtypes(include=object)

X_num.describe()
```

```python
# X_num = fifa_df.drop(['OVA'], axis=1)
```

```python
# MINMAX SCALER NUMERICAL DATA

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

transformer = MinMaxScaler().fit(X_num)
x_normalized = transformer.transform(X_num)
print(x_normalized.shape)
x_normalized = pd.DataFrame(x_normalized, columns=X_num.columns)
x_normalized.head()
```

```python
# ONE HOT ENCODE CATEGORICAL DATA

from sklearn.preprocessing import OneHotEncoder
```

```python
X_cat = ['BP', 'W/F', 'SM', 'A/W', 'D/W' , 'IR']

selected_col = fifa_df[X_cat]

encoder = OneHotEncoder(drop='first').fit(selected_col)
encoded = encoder.transform(selected_col).toarray()

X_cat_encoded = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(X_cat))

X_cat_encoded.head(20)
```

```python
X = pd.concat([x_normalized, X_cat_encoded], axis = 1)
X.describe()
```

```python
# TRAIN TEST SPLIT

from sklearn.model_selection import train_test_split as tts

X_train, X_test, y_train, y_test=tts(X, y, test_size=.1)
```

```python
# CREATE ANOTHER MODEL
# This model will be based on the Best Position of GK (Goalkeeper)
```

```python
# Drop any irrelevant columns that is within other positions
# Filter the BP to only have GK values

fifa_gk = fifa_df.drop(['LS', 'ST', 'RS','LW', 'LF', 'CF', 'RF', 'RW', 'LAM','CAM', 'RAM', 'LM', 'LCM','CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM', 'RWB', 'LB', 'LCB','CB', 'RCB', 'RB'], axis=1)

fifa_gk = fifa_gk[fifa_gk['BP'] == 'GK']

fifa_gk.shape
```

```python
# We still want to base the model on Overall Value

y_gk = fifa_gk['OVA']
X_gk = fifa_gk.drop(['OVA'], axis=1) 

X_num_gk = X_gk.select_dtypes(include=np.number)
X_cat_gk = X_gk.select_dtypes(include=object)

# X_num_gk.describe()


print(X_gk.shape)
print(y_gk.shape)
```

```python
transformer_gk = MinMaxScaler().fit(X_num_gk)
x_normalized_gk = transformer_gk.transform(X_num_gk)
print(x_normalized_gk.shape)
x_normalized_gk = pd.DataFrame(x_normalized_gk, columns=X_num_gk.columns)
x_normalized_gk.head()
```

```python
X_gk.shape
```

```python
X_cat_gk = ['BP', 'W/F', 'SM', 'A/W', 'D/W' , 'IR']

selected_col_gk = X_gk[X_cat_gk]

encoder = OneHotEncoder(drop='first').fit(selected_col_gk)
encoded = encoder.transform(selected_col_gk).toarray()

X_cat_encoded_gk = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(X_cat_gk))

X_cat_encoded_gk.head(20)
```

```python
X_gk = pd.concat([x_normalized_gk, X_cat_encoded_gk], axis = 1)

X_gk.shape

print(x_normalized_gk.shape)
print(X_cat_encoded_gk.shape)
```

```python
from sklearn.model_selection import train_test_split as tts

X_train_gk, X_test_gk, y_train_gk, y_test_gk=tts(X_gk, y_gk, test_size=.25)
```

```python
from sklearn.linear_model import LinearRegression as linreg

lm_gk = linreg()
lm_gk.fit(X_train_gk,y_train_gk)
y_pred_gk = lm_gk.predict(X_test_gk)
```

```python
from sklearn.metrics import r2_score

predictions_gk = lm_gk.predict(X_train_gk)
r2_score(y_train_gk, predictions_gk)
```

```python
predictions_test_gk = lm_gk.predict(X_test_gk)
r2_score(y_test_gk, predictions_test_gk)
```
