## Clean the data

```python
fifa_df = fifa_df.drop(['ID', 'Name', 'Club', 'Hits', 'Nationality', 'Position','Team & Contract','Height','Weight', 'foot','Joined','Loan Date End','Wage','Release Clause','Contract'], axis=1)

fifa_df.columns.values
```

```python
# drop duplicates

fifa_df.drop_duplicates()
```

```python
# nan_positions_tuples = list(zip(*np.where(fifa_df.isna())))
# print(nan_positions_tuples)

fifa_df.columns[fifa_df.isna().any()].tolist()
```

```python
# filling NaN values with mean and mode

volleys_mean = fifa_df['Volleys'].mean()
curve_mean = fifa_df['Curve'].mean()
agility_mean = fifa_df['Agility'].mean()
balance_mean = fifa_df['Balance'].mean()
jumping_mean = fifa_df['Jumping'].mean()
interceptions_mean = fifa_df['Interceptions'].mean()
positioning_mean = fifa_df['Positioning'].mean()
visions_mean = fifa_df['Vision'].mean()
composure_mean = fifa_df['Composure'].mean()
sliding_mean = fifa_df['Sliding Tackle'].mean()
aw_mode = fifa_df['A/W'].mode()[0]
dw_mode = fifa_df['D/W'].mode()[0]


fifa_df['Volleys'].fillna(volleys_mean, inplace=True)
fifa_df['Curve'].fillna(curve_mean, inplace=True)
fifa_df['Agility'].fillna(agility_mean, inplace=True)
fifa_df['Balance'].fillna(balance_mean, inplace=True)
fifa_df['Jumping'].fillna(jumping_mean, inplace=True)
fifa_df['Interceptions'].fillna(interceptions_mean, inplace=True)
fifa_df['Positioning'].fillna(positioning_mean, inplace=True)
fifa_df['Vision'].fillna(visions_mean, inplace=True)
fifa_df['Composure'].fillna(composure_mean, inplace=True)
fifa_df['Sliding Tackle'].fillna(sliding_mean, inplace=True)
fifa_df['A/W'].fillna(aw_mode, inplace=True)
fifa_df['D/W'].fillna(dw_mode, inplace=True)

# check again for any NaN values

fifa_df.columns[fifa_df.isna().any()].tolist()
```

```python
# replace ★ values within columns

fifa_df['W/F'].str.replace('★', '')
fifa_df['SM'].str.replace('★', '')
fifa_df['IR'].str.replace('★', '')
```

```python
def sum_values(value_str):
    values = value_str.split('+')
    return sum(map(int, values))

columns_to_change = ['LS', 'ST', 'RS','LW', 'LF', 'CF', 'RF', 'RW', 'LAM','CAM', 'RAM', 'LM', 'LCM','CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM', 'RWB', 'LB', 'LCB','CB', 'RCB', 'RB', 'GK']

fifa_df[columns_to_change] = fifa_df[columns_to_change].applymap(sum_values)
fifa_df[columns_to_change] = fifa_df[columns_to_change].astype(int)
```

```python
def convert_currency (value):
    if 'K' in value:
        return float(value.replace('€', '').replace('K', '')) * 1000
    elif 'M' in value:
        return float(value.replace('€', '').replace('M', '')) * 1000000
    else:
        return float(value.replace('€', ''))
    
fifa_df['Value'] = fifa_df['Value'].apply(convert_currency)

print(fifa_df['Value'])
```
